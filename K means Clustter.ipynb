```python
X=[15,15,16,19,19,20,20,21,22,28,35,40,41,42,43,44,60,61,65]
import random
#choosing random values from cluster
c1=random.choice(X)
c2=random.choice(X)
flag=True
while flag==True:
    if c1<c2:      #c1 and c2 cannot be same
        flag=False
        break;
    else:
        c1=random.choice(X)
        c2=random.choice(X)
        
print("First random Value :",c1)

print("\nSecond Random Value :",c2)
```
